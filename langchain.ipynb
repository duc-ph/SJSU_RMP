```python
import os
import requests
from dotenv import load_dotenv
load_dotenv()
```

```python
GOOGLE_CSE_ID = os.environ['GOOGLE_CSE_ID']
GOOGLE_API_KEY = os.environ['GOOGLE_API_KEY']
```

```python
def get_answer_or_first_result(query):
    # Define the endpoint and parameters
    url = "https://www.googleapis.com/customsearch/v1"
    params = {
        "key": GOOGLE_API_KEY,
        "cx": GOOGLE_CSE_ID,
        "q": query,
    }

    # Make the request
    response = requests.get(url, params=params)
    results = response.json()

    # Check if there is an answer box in the results
    if "items" in results:
        # Iterate through results to find an answer box
        for item in results["items"]:
            if "pagemap" in item and "answer" in item["pagemap"]:
                return item["pagemap"]["answer"]

        # If no answer box, return the snippet from the first search result
        first_result = results["items"][0]
        if "snippet" in first_result:
            return first_result["snippet"]
        elif "title" in first_result:
            return first_result["title"]

    # If no results at all, return a message
    return "No answer box or search results found."
```

```python
query = 'Yalda Edalat works in which department at sjsu'
get_answer_or_first_result(query)
```
